```python
import pandas as pd
import spacy
from collections import defaultdict

# Carregar o modelo de linguagem do spaCy
!python3 -m spacy download pt_core_news_lg
pln = spacy.load('pt_core_news_lg')
```

```python
# Lista para armazenar os resultados
resultados = []

def contar_adjetivos(texto, contagem_adjetivos):
    doc = pln(str(texto))
    for token in doc:
        if token.pos_ == "ADJ" or (token.pos_ == "VERB" and token.morph.get("VerbForm") == ["Part"]):
            contagem_adjetivos[token.text] += 1

#carregue o DataFrame aqui
dt = pd.read_csv('/content/drive/MyDrive/TCC - UFMS/CSBC-WICS/Resultados/Frases/resultados_arvore_dependencia_feminino.csv')

# Verificar e remover os valores nulos na coluna 'Frase'
dt = dt.dropna(subset=['Frase'])
total_frases_apos_remocao_nulos = len(dt)
print(f"Total de frases após remoção de nulos: {total_frases_apos_remocao_nulos}")
# Dicionário para armazenar a contagem de adjetivos
contagem_adjetivos = defaultdict(int)

# Contagem de adjetivos em cada texto do conjunto de dados
for texto in dt['Frase']:
    contar_adjetivos(texto, contagem_adjetivos)

# Classificar os adjetivos por contagem em ordem decrescente
ranking_adjetivos = sorted(contagem_adjetivos.items(), key=lambda x: x[1], reverse=True)

# Imprimir o ranking de adjetivos
print("Ranking de Adjetivos:")
for i, (adjetivo, contagem) in enumerate(ranking_adjetivos, start=1):
    resultados.append({'Adjetivo/Verbo no Particípio': adjetivo, 'Contagem': contagem})

print(f"Quantidade de adjetivos distintos: {len(contagem_adjetivos)}")
# Converter os resultados em um DataFrame e exibir as primeiras linhas
resultados_df = pd.DataFrame(resultados)
print(resultados_df.head())
```
